```python
# =============================================
# STEP 1: Import Libraries & Load Data
# =============================================

import pandas as pd                  # For data manipulation
import matplotlib.pyplot as plt      # For charts
import seaborn as sns                # For beautiful charts

# Load the dataset
df = pd.read_csv("Pro/laptop_sales.csv")

# ---- EXPLORE THE DATA ----

print("Shape (rows, columns):", df.shape)        # How big is the data?
print("\nColumn Names:\n", df.columns.tolist())   # What columns exist?
print("\nFirst 5 Rows:")
print(df.head())                                  # Preview first 5 rows
print("\nData Types:")
print(df.dtypes)                                  # What type is each column?
print("\nBasic Statistics:")
print(df.describe())                              # Min, max, mean of numbers
```

```python
# =============================================
# PART 1 — DATA MANIPULATION TASKS
# =============================================

# ✅ Task 1: Total Revenue
total_revenue = df["Revenue"].sum()
print("1️⃣ Total Revenue:", total_revenue)

# ✅ Task 2: Total Profit
total_profit = df["Profit"].sum()
print("2️⃣ Total Profit:", total_profit)

# ✅ Task 3: Total Quantity Sold
total_quantity = df["Quantity"].sum()
print("3️⃣ Total Quantity Sold:", total_quantity)

# ✅ Task 4: Average Unit Price
avg_price = df["Unit_Price"].mean()
print("4️⃣ Average Unit Price:", round(avg_price, 2))

# ✅ Task 5: Top 5 Brands by Revenue
top_brands = df.groupby("Brand")["Revenue"].sum().sort_values(ascending=False).head(5)
print("\n5️⃣ Top 5 Brands by Revenue:")
print(top_brands)
```

```python
# ✅ Task 6: Most Sold Laptop Type
most_sold_type = df.groupby("Laptop_Type")["Quantity"].sum().sort_values(ascending=False)
print("6️⃣ Most Sold Laptop Type:")
print(most_sold_type)

# ✅ Task 7: City with Highest Sales
top_city = df.groupby("City")["Revenue"].sum().sort_values(ascending=False).head(1)
print("\n7️⃣ City with Highest Revenue:")
print(top_city)

# ✅ Task 8: State with Highest Profit
top_state = df.groupby("State")["Profit"].sum().sort_values(ascending=False).head(1)
print("\n8️⃣ State with Highest Profit:")
print(top_state)

# ✅ Task 9: Monthly Revenue Trend
df["Date"] = pd.to_datetime(df["Date"])         # Convert Date column to date format
df["Month"] = df["Date"].dt.to_period("M")      # Extract year-month (e.g. 2023-01)
monthly_revenue = df.groupby("Month")["Revenue"].sum()
print("\n9️⃣ Monthly Revenue Trend:")
print(monthly_revenue)

# ✅ Task 10: Profit Margin Column
df["Profit_Margin"] = (df["Profit"] / df["Revenue"]) * 100
avg_margin = df["Profit_Margin"].mean()
print(f"\n🔟 Average Profit Margin: {round(avg_margin, 2)}%")
```

```python
# =============================================
# PART 2 — VISUALIZATION TASKS
# =============================================

# ✅ Task 1: Bar Chart — Revenue by Brand
brand_revenue = df.groupby("Brand")["Revenue"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=brand_revenue.index, y=brand_revenue.values, palette="Blues_d")
plt.title("Revenue by Brand")
plt.xlabel("Brand")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ Task 2: Pie Chart — Laptop Type Distribution
type_counts = df.groupby("Laptop_Type")["Quantity"].sum()

plt.figure(figsize=(7, 7))
plt.pie(type_counts.values, labels=type_counts.index, autopct="%1.1f%%", startangle=140)
plt.title("Laptop Type Distribution")
plt.show()

# ✅ Task 3: Bar Chart — Top 10 Cities by Revenue
top_cities = df.groupby("City")["Revenue"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_cities.index, y=top_cities.values, palette="Greens_d")
plt.title("Top 10 Cities by Revenue")
plt.xlabel("City")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ Task 4: Line Chart — Monthly Revenue Trend
plt.figure(figsize=(12, 5))
monthly_revenue.plot(kind="line", marker="o", color="blue")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ Task 5: Bar Chart — Profit by Brand
brand_profit = df.groupby("Brand")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=brand_profit.index, y=brand_profit.values, palette="Oranges_d")
plt.title("Profit by Brand")
plt.xlabel("Brand")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ✅ Task 1: Bar Chart — Revenue by Brand (FIXED)
brand_revenue = df.groupby("Brand")["Revenue"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=brand_revenue.index, y=brand_revenue.values, 
            hue=brand_revenue.index, palette="Blues_d", legend=False)
plt.title("Revenue by Brand")
plt.xlabel("Brand")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ Task 3: Top 10 Cities by Revenue (FIXED)
top_cities = df.groupby("City")["Revenue"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_cities.index, y=top_cities.values, 
            hue=top_cities.index, palette="Greens_d", legend=False)
plt.title("Top 10 Cities by Revenue")
plt.xlabel("City")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ Task 5: Profit by Brand (FIXED)
brand_profit = df.groupby("Brand")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=brand_profit.index, y=brand_profit.values, 
            hue=brand_profit.index, palette="Oranges_d", legend=False)
plt.title("Profit by Brand")
plt.xlabel("Brand")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ✅ Task 6: Scatter Plot — Unit Price vs Profit
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Unit_Price"], y=df["Profit"], alpha=0.5, color="purple")
plt.title("Unit Price vs Profit")
plt.xlabel("Unit Price")
plt.ylabel("Profit")
plt.tight_layout()
plt.show()

# ✅ Task 7: Histogram — Distribution of Unit Price
plt.figure(figsize=(10, 5))
sns.histplot(df["Unit_Price"], bins=30, kde=True, color="steelblue")
plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# ✅ Task 8: Box Plot — Revenue by Laptop Type
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Laptop_Type"], y=df["Revenue"], palette="Set2")
plt.title("Revenue by Laptop Type")
plt.xlabel("Laptop Type")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ Task 9: Heatmap — Correlation Matrix
plt.figure(figsize=(8, 6))
cols = ["Quantity", "Unit_Price", "Unit_Cost", "Revenue", "Profit"]
correlation = df[cols].corr()
sns.heatmap(correlation, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()

# ✅ Task 10: Top 10 Models by Quantity Sold
top_models = df.groupby("Model_Name")["Quantity"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_models.index, y=top_models.values,
            hue=top_models.index, palette="Purples_d", legend=False)
plt.title("Top 10 Models by Quantity Sold")
plt.xlabel("Model Name")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
